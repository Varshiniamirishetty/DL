```python
#hebbian learning model
import numpy as np

# Convert I / O pattern to bipolar (+1 / -1)
def convert(pattern):
    return np.array([1 if ch == 'I' else -1 for ch in pattern])

# -------------------------------
# Define input patterns
# -------------------------------

# Pattern 1 (y = +1)
p1 = [
    'I','I','I',
    'O','I','O',
    'I','I','I'
]

# Pattern 2 (y = -1)
p2 = [
    'O','I','O',
    'I','O','I',
    'O','I','O'
]

# Convert patterns to bipolar form
x1 = convert(p1)
x2 = convert(p2)

# Desired outputs
y1 = 1
y2 = -1

# Learning rate
eta = 1

# Initialize weight vector
w = np.zeros(len(x1))
b=np.zeros(1)
# -------------------------------
# Hebbian Learning (pattern-based)
# -------------------------------
w = w + eta * x1 * y1
b=b+y1
print("w1",w,"b1",b)
w = w + eta * x2 * y2
b=b+y2
print("baised",b)

# Display final weights
print("Final weight vector:")
print(w)

# -------------------------------
# Pattern Recall
# -------------------------------
def recall(x, w,b):
    net = np.dot(x, w)+b
    return 1 if net >= 0 else -1

print("\nRecall Pattern 1:", recall(x1, w,b))
print("Recall Pattern 2:", recall(x2, w,b))
```

```python
#multilayered perceptron model
import numpy as np

# Input patterns
x  = [-1, -1,  1,  1]
x1 = [-1,  1, -1,  1]

# Target outputs
t = [-1, 1, 1, 1]

# Initialize weights and bias
w1 = 0
w2 = 0
b = 0

eta = 1

# Training
for i in range(len(x)):
    net = w1*x[i] + w2*x1[i] + b

    if net != t[i]:
        w1 = w1 + eta * x[i]  * t[i]
        w2 = w2 + eta * x1[i] * t[i]
        b  = b  + eta * t[i]
    print("Final weights and bias:")
    print("w1 =", w1, "w2 =", w2, "b =", b)

```

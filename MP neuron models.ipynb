```python
#AND,OR,NOT_AND,XOR
#n-number of excitatory values required
#w-distinct weights of excitatory values
#p-number of inhibitory values
def mp_neuron(x1, x2, w1, w2, t):
    y_in = x1 * w1 + x2 * w2
    if y_in >= t:
        y = 1
    else:
        y = 0
    return y_in, y


def calculate_threshold(n, w, p):
    return n * w - p


print("AND Gate")
n, w, p = 2, 1, 0
t = calculate_threshold(n, w, p)
w1, w2 = 1, 1

for x1 in [0, 1]:
    for x2 in [0, 1]:
        y_in, y = mp_neuron(x1, x2, w1, w2, t)
        print(f"x1={x1}, x2={x2}, y_in={y_in}, y={y}")


print("\nOR Gate")
n, w, p = 1,1,0
#here n=1 because 1 excitatory value is enough
t = calculate_threshold(n, w, p)
w1, w2 = 1,1

for x1 in [0, 1]:
    for x2 in [0, 1]:
        y_in, y = mp_neuron(x1, x2, w1, w2, t)
        print(f"x1={x1}, x2={x2}, y_in={y_in}, y={y}")


print("\nAND-NOT Gate (x1 AND NOT x2)")
n, w, p = 2, 1, 1
t = calculate_threshold(n, w, p)
w1, w2 = 1, -1

for x1 in [0, 1]:
    for x2 in [0, 1]:
        y_in, y = mp_neuron(x1, x2, w1, w2, t)
        print(f"x1={x1}, x2={x2}, y_in={y_in}, y={y}")


print("\nXOR Gate")

for x1 in [0, 1]:
    for x2 in [0, 1]:
        # Pre-compute terms
        X = x1 * (1 - x2)   # x1 AND NOT x2
        Y = (1 - x1) * x2   # NOT x1 AND x2

        # Weights
        w1, w2 = 1, 1

        # Threshold for OR
        t = calculate_threshold(1, 1, 0)  # threshold = 1

        # Single MP neuron call
        y_in, y = mp_neuron(X, Y, w1, w2, t)

        print(f"x1={x1}, x2={x2}, y_in={y_in}, y={y}")
```

```python
#ACTIVATION FUNCTIONS
def relu(x):
    return np.maximum(0, x)

def lrelu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softplus(x):
    return np.log1p(np.exp(x))

def tanh(x):
    return np.tanh(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
```

```python
plt.figure()
plt.plot(x, relu(x))
plt.title("ReLU")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.plot(x, lrelu(x))
plt.title("Leaky ReLU")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
-+plt.figure()
plt.plot(x, sigmoid(x))
plt.title("Sigmoid")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.plot(x, softplus(x))
plt.title("Softplus")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.plot(x, tanh(x))
plt.title("Tanh")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)

x_sm = np.array([-2, -1, 0, 1, 2])
y_sm = softmax(x_sm)

plt.figure()
plt.bar(range(len(x_sm)), y_sm)
plt.title("Softmax")
plt.xlabel("Index")
plt.ylabel("Probability")
plt.show()
```
